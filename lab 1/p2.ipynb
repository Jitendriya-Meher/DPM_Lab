 Load  the  breast  cancer  dataset  from  the  UCI  machine  learning 
repository using Pandas.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("BC.csv")
df
```

Replace  missing  values  (encoded  as  '?')  with  NaNs  in  the  breast 
cancer dataset.

```python
df.replace('?', np.nan, inplace=True)

df
```

Count the number of missing values in each column of the breast 
cancer dataset?

```python
df.isnull().sum()
```

Replace  missing  values  in  the  'Bare  Nuclei'  column  with  the 
median value of that column.

```python
bData = []

for i in (df["Bare_Nuclei"]):
    # print(i,type(i))
    bData.append(float(i))

# print(df["Bare_Nuclei"].median())

# print("\nNull cols :\n",df.isnull().sum())

bData = pd.DataFrame(data=bData)

med = bData.median()

df.replace(np.nan,med[0], inplace=True)
df
```
