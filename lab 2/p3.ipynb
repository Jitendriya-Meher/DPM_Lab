```python
import pandas as pd
import matplotlib.pyplot as plt
```

 Load  the  breast  cancer  dataset  from  the  UCI  machine  learning 
repository using Pandas

```python
df = pd.read_csv("BC_data.csv")
df
```

Draw  a  boxplot  to  identify  columns  with  outliers  in  the  breast

```python
# boxplot
# for col in df:
#     plt.boxplot(df[col])
#     plt.title(col)
#     plt.show()
```

```python
newDf = df.drop(columns=["ID"])
# newDf
newDf.boxplot()
plt.xticks(rotation=90)
plt.show()
```

Standardize the columns of the breast cancer dataset to compute 
the Z-scores

```python
# z score 
z = (newDf - newDf.mean(numeric_only=True)) / newDf.std(numeric_only=True)
print("z score =",z)
```

Identify and count duplicate rows in the breast cancer dataset. 


```python
# count and show duplicate
isDubRows = df.duplicated()

count=0
for i in  range (len(isDubRows)):
    if( isDubRows[i]==True):
        count+=1
        print( df.loc[i])
    
print("\nTotal duplicate rows :", count)
```

Remove duplicate rows from the breast cancer dataset

```python
# remove duplicate 
newDf = df.drop_duplicates()
newDf
```

Randomly select a sample of size 3 (without replacement) from the 
breast cancer dataset.

```python
# select a sample of df 
sample_df = df.sample(n=3, replace=False)
sample_df
```

Randomly  select  1%  of  the  data  (without  replacement)  from  the 
breast cancer dataset

```python
n = int(len(df)*0.01)
print("n =",n)
sample_df = df.sample(n, replace=False)
sample_df
```

Perform sampling with replacement to create a sample whose size 
is equal to 1% of the entire breast cancer dataset. 

```python
n = int(len(df)*0.01)
print("n =",n)
sample_df = df.sample(n, replace=True)
sample_df
```
