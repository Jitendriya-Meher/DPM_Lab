```python
df = pd.read_csv("BC_data.csv")
df
```

```python
newDf = df.drop(columns=["ID"])
# newDf
newDf.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
# z score 
z = (newDf - newDf.mean(numeric_only=True)) / newDf.std(numeric_only=True)
print("z score =",z)
```

```python
# count and show duplicate
isDubRows = df.duplicated()

count=0
for i in  range (len(isDubRows)):
    if( isDubRows[i]==True):
        count+=1
        print( df.loc[i])
    
print("\nTotal duplicate rows :", count)
```

```python
# remove duplicate 
newDf = df.drop_duplicates()
newDf
```
