```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("data_banknote_authentication.txt")
data
```

```python
cols = ['variance','skewness','curtosis','entropy','class']

data.columns = cols
data
```

```python
feature = data[['variance','skewness','curtosis','entropy']]
feature
```

```python
clas = data[['class']]
clas = np.array(clas)
clas = clas.reshape(-1)
clas
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature,clas, test_size=0.2, random_state=42)

X_train, X_test, y_train, y_test
```

```python
knnModel = KNeighborsClassifier(n_neighbors=5)
knnModel
```

```python
knnModel.fit(X_train,y_train)
```

```python
ipvariance = int(input("Enter the Variance"))
ipskewness = int(input("Enter the Skewness"))
ipcurtosis = int(input("Enter the curtosis"))
ipentropy = int(input("Enter the entropy"))

ip = [[ipvariance,ipskewness,ipcurtosis,ipentropy]]
ip
```

```python
opClass = knnModel.predict(ip)
opClass
```

```python
if opClass[0] == 0:
    print("No")
else:
    print("Yes")
```
